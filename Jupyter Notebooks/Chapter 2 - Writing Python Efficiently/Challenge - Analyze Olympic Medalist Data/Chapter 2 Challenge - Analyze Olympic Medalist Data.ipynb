```python
#Create the collection and load the data into it
medal = collections.namedtuple('medal',['year','athlete','team','event']) # namedtuple(objectName,[objectAttribute])
medals = [medal(*line.strip().split('\t')) for line in open('goldmedals.txt','r')] 
```

## Challenge 1: <br>  Find the five athletes who won the most gold medals

```python
def most_gold_medals(medals):
    return collections.Counter(medal.athlete for medal in medals).most_common(5)
```

```python
for athlete, count in most_gold_medals(medals):
    print("Athlete {0} has won {1} medal(s)." .format(athlete, count))
```

## Challenge 2: <br>  Find the five athletes who won gold in the largest number of different events

```python
events_by_athlete_set = collections.defaultdict(set) #default dict set to help avoid repition
for medal in medals: 
        events_by_athlete_set[medal.athlete].add(cleaneventName(medal.event))  
               
def howmany(tup):
    return len(tup[1])

def cleaneventName(event):
    event.split()
    return ''.join(word for word in event.split() if word not in ('men', 'women'))
    
sorted(events_by_athlete_set.items(),key = howmany, reverse = True)
```
