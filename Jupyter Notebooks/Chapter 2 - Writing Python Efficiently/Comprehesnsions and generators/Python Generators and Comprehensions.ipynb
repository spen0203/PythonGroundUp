#### Dictionary Simplified:

```python
cities_by_year = {year: city for city,year in zip(cities,years)}
# dictionary =   {sort key: elemenent, loop }
# {'1896': 'Athens',
#  '1900': 'Paris',
```

```python
cities_by_year
```

```python
#Create a set based off cities after 1930
cities_after_1930 = { city for year,city in cities_by_year.items() if int(year) > 1930}
cities_after_1930
```

```python
#Draw map using the new set
pp.figure(figsize=(8,8))
world =Basemap(projection='ortho', lat_0=75, lon_0=0)
world.drawcoastlines(linewidth =0.25)
world.drawcountries(linewidth = 0.25)

#loop over cities
for year,city in cities_by_year.items():
    x,y = world(*coordinates_by_city[city]) # world() takes two params * wildcardss it
    
    world.plot(x,y,'r.') # add a dot to the map
    pp.text(x,y,year,fontsize=8)
```

# Comprehensions - END  <br> Generator Expressions - START
<p> 
Generators, create a generator object, theese are different from a list in the fact it deals with one element at a time but can be used with a function in the same way a list would. <b>Also are significantly faster and smaller than a list  BUT you can only iterate over once.  </b> 
<br>they can be noticed by () ie: <ul> <li> even = ( i for i in range(20) if i %2 == 0) </li>
   <li>[next(f) for i in range(20)] # __next__() shorthand, called 20 times </li></ul>    
</p>

```python
even = ( i for i in range(20) if i % 2 == 0) #Create a generator object
```

```python
even.__next__() #iteratable
```

```python
even.__next__()
```

```python
sum(even) # can pass generator objects to functions similiar to a list
```

### Use keyword - yield , to return a generator

```python
def fibonacci():
    f1,f2 = 0, 1 #init vars
    while True: #infinite loop
        yield f2 # returns f2 as a generator
        f1, f2 = f2, f1 + f2 #f1 = f2 , f2 = f1 + f2 - done in one line
```

```python
f = fibonacci()
```

```python
[next(f) for i in range(20)] # __next__() shorthand, called 20 times
```
