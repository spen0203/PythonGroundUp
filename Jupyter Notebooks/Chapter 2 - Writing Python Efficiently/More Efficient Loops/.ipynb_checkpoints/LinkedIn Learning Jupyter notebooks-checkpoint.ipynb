```python
# prints a from file [ City , Year ]
# Fails for citys with multiple words
for game in open('games.txt','r'):
    city = game.split()[0]
    year = game.split()[1]
    
    print(city,year)
```

```python
# prints from a file [ City , Year ]
# Fixed for citys with multiple words, better ways to work with a list
for game in open('games.txt','r'):
    words = game.split() #split the game.txt line into a list of words
    
    city = ' '.join(words[:-1]) # city joins each item from the words list except the last item. (year)
    year = words[-1].strip('()') # words[-1], another way to get last list item, cool python feature
    print(city,year)
```

```python
# lists from a file [ City , Year ] 

cities, years =[] , []
# Fixed for citys with multiple words, better ways to work with a list
for game in open('games.txt','r'):
    words = game.split() #split the game.txt line into a list of words
    
    city = ' '.join(words[:-1]) # city joins each item from the words list except the last item. (year)
    year = words[-1].strip('()') # words[-1], another way to get last list item, cool python feature
    
    cities.append(city)
    years.append(year)
    
```

```python
# Check to see if its already installed
!pip install geopy 
import geopy


geolocator = geopy.geocoders.Nominatim(user_agent="my-application") #geolocator object

locations = {} #dictionary
for city in cities: #Jupyter Notebooks allows you to share vars between code cells(each cells kinda its own function)
    print("Locating ", city)
    locations[city] = geolocator.geocode(city.split('/')[0]) #Updates the location dictionary 
print("Geolocator: \n")
print(locations["Paris"])
```

```python
#Plot each city on the map
#Uses "import matplotlib.pyplot as pp" from cell 1

pp.figure(figsize=(10,5))
world = Basemap() #create map object
world.drawcoastlines(linewidth=0.25) #draw details
world.drawcountries(linewidth=0.25)

for city,pos in locations.items():#for each item in the locations dictionary
    world.plot(pos.longitude, pos.latitude,'r.',markersize=10, latlon=True) #plot the city with red .
```

```python
#Fix two locations to be more specific
locations['Rome'] = geolocator.geocode('Rome, Italy')
locations['Athens'] = geolocator.geocode('Athens, Greece')


pp.figure(figsize=(10,5))
world = Basemap() #create map object
world.drawcoastlines(linewidth=0.25) #draw details
world.drawcountries(linewidth=0.25)

for city,pos in locations.items():#for each item in the locations dictionary
    world.plot(pos.longitude, pos.latitude,'r.',markersize=10, latlon=True) #plot the city with red .
```

```python
#First 10 items of list
for i, city in enumerate(cities[:10]):
    print(i,city)
```

```python
#Sorted List of 10 items
for city in sorted(cities[:10]):
    print(city)
```

```python
#Sorted List based on string length of 10 items
for city in sorted(cities[:10], key=len): 
    print(city)
```

```python
#Last 10 items of a list
for i, city in enumerate(reversed(cities[:10])):
    print(i,city)
```

```python
#Work jointly on iterator, ends when shortest condition is met
for year, city in zip(years[:10], cities): #First 10 year items & all corresponding cities
    print(year,city)
```
